```python
import numpy as np
import math
from random import randrange
from matplotlib import pyplot as plt

x = np.array([])
y = np.array([])
MRrange = np.array([])
RLCUSUM = np.array([])
RLIX = np.array([])

target_mean = 10
mean = 10
sigma = 1
run = 1000 # Number of ARL runs (Ideally 1,000)
p = 600 # Number of variables in X (Ideally 10,000)
n = 2 # Sample Size
K = 0.5 * sigma
H = 5 * sigma
d2 = 1.128

z = np.zeros(p-1)
Cplus = np.zeros(p-1)

seed = 0
i = 0
j = 0

for i in range (run):
    print (i)
    X = np.random.normal(mean, sigma, p)
    Xbar = np.average(X)
    
     
    ############# IX #############
    for j in range(p):
        MRi = abs(X[j]-X[j-1])
        MRrange = np.append(MRrange,MRi)
    
    MRrange = np.delete(MRrange,0)
    MRbar = np.average(MRrange)
    
    UCL = Xbar + sigma * (MRbar / d2)
    #CenterLine = Xbar
    #LCL = Xbar - sigma * (MRbar / d2)
    
    ############# Count for IX #############
    b = 0
    while X[b] < UCL:
        b = b + 1
        if(b == (p-1)):
            break
   
    RLIX = np.append(RLIX,b)
    
    ############# CUMSUM #############
    for j in range (p-1):

        C = np.maximum(z[j],(X[j]-(target_mean + K) + Cplus[j-1]))
        Cplus[j] = C

    
    ############# Count for CUSUM #############
    b = 0
    while Cplus[b] < H:
        b = b + 1
        if(b == (p-1)):
            break  
            
    RLCUSUM = [number for number is Cplus if number % b<]        
            
    RLCUSUM = np.append(RLCUSUM,b)

RLIXbar = np.average(RLIX)
print("\n RLIXbar is:")
print (RLIXbar) 
    
RLCUSUMbar = np.average(RLCUSUM)
print("\n RLCUSUMbar is:")
print (RLCUSUMbar)
```
